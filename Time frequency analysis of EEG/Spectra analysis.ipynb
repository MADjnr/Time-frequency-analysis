```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this is a code for auto regression, in the simplest form,
#it is like regressing varaible on its self. 

alpha = np.array([-.3, .8])

k = len(alpha)
N = 200
x = np.zeros(N)

for i in range(k,N):
    x[i] = sum(alpha*x[i-k:i]) + np.random.randn()
    
    
x += np.sin(np.linspace(0,10*np.pi, N))*10  
    
plt.plot(x)
plt.xlabel('X axis')
plt.title('AR(%s) process' %k)
plt.show()
```
